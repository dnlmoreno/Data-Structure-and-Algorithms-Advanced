```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tamaños = pd.read_csv('../Resultados/Tamaños_medianos_datos/Experimentacion_busquedas_tamanios.csv', sep = ';')
posiciones = pd.read_csv('../Resultados/Tamaños_medianos_datos/Experimentacion_busquedas_posicion.csv', sep = ';')

tamaños_v1 = pd.read_csv('../Resultados/Tamaños_grandes_datos/Experimentacion_busquedas_tamanios.csv', sep = ';')
posiciones_v1 = pd.read_csv('../Resultados/Tamaños_grandes_datos/Experimentacion_busquedas_posicion.csv', sep = ';')
```

```python
col0_tam = tamaños['Tamaños']
col1_tam = tamaños['Tiempo_búsqueda_lineal']
col2_tam = tamaños['Tiempo_búsqueda_binaria']
col3_tam = tamaños['Tiempo_búsqueda_doblada']

col0_tam_v1 = tamaños_v1['Tamaños']
col1_tam_v1 = tamaños_v1['Tiempo_búsqueda_binaria']
col2_tam_v1 = tamaños_v1['Tiempo_búsqueda_doblada']
```

```python
posiciones
```

```python
posiciones_v1   
```

```python
col0_pos = posiciones['Posición']
col1_pos = posiciones['Tiempo_búsqueda_lineal']
col2_pos = posiciones['Tiempo_búsqueda_binaria']
col3_pos = posiciones['Tiempo_búsqueda_doblada']

col0_pos_v1 = posiciones_v1['Posición']
col1_pos_v1 = posiciones_v1['Tiempo_búsqueda_binaria']
col2_pos_v1 = posiciones_v1['Tiempo_búsqueda_doblada']
```

```python
fig, ax1 = plt.subplots(figsize=(11,8))

ax1.set_xlabel('Tamaño de la entrada', fontsize='20', labelpad=15)
ax1.set_ylabel('Nanosegundos', fontsize='20', labelpad=16)
ax1.plot(col0_tam, col1_tam, marker="o", color= "c", label='Búsqueda lineal', linewidth=2.5, markersize=7)
ax1.plot(col0_tam, col2_tam, marker="o", color= "y", label='Búsqueda binaria', linewidth=2.5, markersize=7)
ax1.plot(col0_tam, col3_tam, marker="o", color= "tab:purple", label='Búsqueda doblada', linewidth=2.5, markersize=7)
ax1.tick_params(axis='y')

plt.xticks(fontsize = 15)
plt.yticks(size = 15)

## Agranda tamaño del 1e6 del eje x
#ax1.xaxis.get_offset_text().set_fontsize(17)
#ax1.xaxis.get_offset_text().set_x(1.05)

legend = plt.legend(loc='upper left', shadow=True, fontsize='15')
legend.get_frame().set_facecolor('white')

plt.savefig("Grafico_tamaños_v0.png", bbox_inches='tight')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(11,8))

ax1.set_xlabel('Tamaño de la entrada', fontsize='20', labelpad=15)
ax1.set_ylabel('Nanosegundos', fontsize='20', labelpad=16)
ax1.plot(col0_tam_v1, col1_tam_v1, marker="o", color= "y", label='Búsqueda binaria')
ax1.plot(col0_tam_v1, col2_tam_v1, marker="o", color= "tab:purple", label='Búsqueda doblada')
ax1.tick_params(axis='y')

plt.xticks(fontsize = 15)
plt.yticks(size = 15)

## Agranda tamaño del 1e6 del eje x
ax1.xaxis.get_offset_text().set_fontsize(17)
ax1.xaxis.get_offset_text().set_x(1.05)

legend = plt.legend(loc='upper left', shadow=True, fontsize='15')
legend.get_frame().set_facecolor('white')

plt.savefig("Grafico_tamaños_v1.png", bbox_inches='tight')
plt.show()
```

```python
''' Codigo servible en otros experimentos '''

fig, (ax, ax2) = plt.subplots(2, 1, sharex=True, figsize=(11,8))

''' plot the same data on both axes '''
# Parte de abajo del grafico
ax2.plot(col0_tam, col1_tam, marker="o", color= "c", label='Búsqueda lineal', linewidth=2.5, markersize=7)
ax2.plot(col0_tam, col2_tam, marker="o", color= "y", label='Búsqueda binaria', linewidth=2.5, markersize=7)
ax2.plot(col0_tam, col3_tam, marker="o", color= "tab:purple", label='Búsqueda doblada', linewidth=2.5, markersize=7)

# Parte de arriba del grafico
ax.plot(col0_tam, col1_tam, marker="o", color= "c", label='Búsqueda lineal', linewidth=2.5, markersize=7)
ax.plot(col0_tam, col2_tam, marker="o", color= "y", label='Búsqueda binaria', linewidth=2.5, markersize=7)
ax.plot(col0_tam, col3_tam, marker="o", color= "tab:purple", label='Búsqueda doblada', linewidth=2.5, markersize=7)
ax.plot(col0_tam, col3_tam, marker="o", color= "w", linewidth=3.8, markersize=8) # El HAAACK

# zoom-in / limit the view to different portions of the data
ax.set_ylim(5000, 832335)  # outliers only
ax2.set_ylim(0, 1200)  # most of the data

# hide the spines between ax and ax2
ax.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

## Agranda los ticks de los ejes x e y con un grafico con ejes broke
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(15) 
plt.xticks(fontsize = 15)
plt.yticks(size = 15)

ax2.set_xlabel('Tamaño de la entrada', fontsize='20', labelpad=13)
ax2.set_ylabel('Nanosegundos', fontsize='20')
ax2.yaxis.set_label_coords(0.025, 0.5, transform=fig.transFigure)

ax.legend(loc='upper left', shadow=True, fontsize='15')
#ax2.legend(loc='upper left', shadow=True, fontsize='15')

d = .015  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

#plt.savefig("Grafico_tamaños_v0.png", bbox_inches='tight')
plt.show()
```
